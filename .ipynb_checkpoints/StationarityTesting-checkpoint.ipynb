```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.arima_model import ARIMA
from pandas.plotting import register_matplotlib_converters
from constants import macroWords, sectorWords, tickerList, companies
tickerDict = {v: k for k, v in companies.items()}
```

```python
df = pd.read_csv('Data/stocks_dateRange/googl.csv', parse_dates=['date'], index_col=['date'])
```

```python
df.plot(y='adjclose', use_index=True)
df.head()
```

```python
rolling_mean = df['adjclose'].rolling(window=12).mean()
rolling_std = df['adjclose'].rolling(window=12).std()

plt.plot(df.index, df['adjclose'], color='blue', label='Original')
plt.plot(rolling_mean, color = 'red', label = 'Rolling Mean')
plt.plot(rolling_std, color = 'black', label ='Rolling Std')
plt.legend(loc = 'best')
plt.title('Rolling Mean and Rolling Standard Deviation')
plt.show()

# Notice from the plot below that the standard deviation is stationary
```

```python
def get_stationarity(data, col):
    rolling_mean = data[col].rolling(window=12).mean()
    rolling_std = data[col].rolling(window=12).std()

    plt.plot(data.index, data[col], color='blue', label='Original')
    plt.plot(rolling_mean, color = 'red', label = 'Rolling Mean')
    plt.plot(rolling_std, color = 'black', label ='Rolling Std')
    plt.legend(loc = 'best')
    plt.title('Rolling Mean and Rolling Standard Deviation')
    plt.show(block=False)
    
    # Dickey-Fuller test
#     result = adfuller(data[col])
#     print('ADF Statistic: {}'.format(result[0]))
#     print('p-value: {}'.format(result[1]))
#     print('Critical Values:')
#     for key, value in result[4].items():
#         print('\t{}:{}'.format(key, value))
    
```

```python
rolling_mean = df['adjclose'].rolling(window=12).mean()
rolling_std = df['adjclose'].rolling(window=12).std()

df_minus_mean['adjclose'] = df['adjclose'] - rolling_mean

get_stationarity(df_minus_mean, 'adjclose')
```

```python
df_shift = df - df.shift()

get_stationarity(df_shift, 'adjclose')
```

```python
df_shift2 = df_shift - df_shift.shift()
get_stationarity(df_shift2, 'adjclose')
```
